Scenario 5: Medicine Effectiveness Study
A company made a new medicine to lower blood pressure. They gave it to one group and gave a fake pill (placebo) to another group.
Question 1:
 How can the company check if the new medicine works?


```python
import numpy as np
from scipy.stats import ttest_ind

bp_before_medicine = np.array([150, 145, 160, 155, 148])
bp_before_placebo  = np.array([149, 146, 158, 154, 150])

bp_after_medicine = np.array([130, 128, 135, 132, 129])
bp_after_placebo  = np.array([145, 143, 150, 148, 147])

reduction_medicine = bp_before_medicine - bp_after_medicine
reduction_placebo  = bp_before_placebo - bp_after_placebo

print("Average BP reduction (Medicine):", reduction_medicine.mean())
print("Average BP reduction (Placebo):", reduction_placebo.mean())

t_stat, p_value = ttest_ind(reduction_medicine, reduction_placebo)

print("P-value:", p_value)

if p_value < 0.05:
    print("Medicine is EFFECTIVE")
else:
    print("Medicine is NOT effective")
```

Scenario 6: Identifying Outliers in Sales Data
A company wants to find any unusual spikes in sales.
Question 1:
 How can the company detect outliers in their sales data?


```python
import pandas as pd
sales=pd.Series([200,220,210,215,230,1000])
Q1=sales.quantile(0.25)
Q3=sales.quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
outliers=sales[(sales<lower)| (sales>upper)]
print("lower",lower)
print("upper",upper)
print("outliers:\n",outliers)
```
