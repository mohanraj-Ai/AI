```python
import pandas as pd #pandas for table usage and convert pandas to pd
```

```python
dataset=pd.read_csv("Salary_Data.csv")#in class pd execute the read function and call the csv file
```

```python

dataset # run the dataset variable
```

```python
independent=dataset[["YearsExperience"]] # in Model creation, this is the input split-up process. in here independent is input variable. in this code to take the required column from the dataset.
independent                
```

```python
dependent=dataset[["Salary"]] # in the model creation, this is output split-up process. in here dependent worked as a output. in this code, to get output column from the dataset.
dependent
```

```python
from sklearn.model_selection import train_test_split #In model creation, this is split sk learn library used in ML. In model creation, this is split the train and set process. in here i called the function train_test_split from the sklearn(library) folder model_selectio.
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression # in here get the LinearRegression class from the sklearn libray 
regressor=LinearRegression() # renamed the variable name for our purpose and function assigned this name
regressor.fit(X_train,y_train) #in regressor to call fit function to subtitude the train set values
```

```python
weight=regressor.coef_ # to find the weight value. in the code using coef method to findout the weight value
weight
```

```python
 bais=regressor.intercept_ # to find the bais value using intercept function
bais
```

```python
y_pred=regressor.predict(X_test) #test set evalution process
```

```python
from sklearn.metrics import r2_score # from sklearn library get metricts function and r2 score fuction, to find the r2 vale
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle #pickle fuction for save the model
filename="finalized_model_linear.sav" # to assign variable name
```

```python
pickle.dump(regressor,open(filename,'wb'))  # in pickle dump our model 
```
