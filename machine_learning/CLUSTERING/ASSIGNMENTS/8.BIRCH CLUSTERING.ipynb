```python
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import Birch
```

```python
classifier=Birch( threshold=0.5, branching_factor=50, n_clusters=3, compute_labels=True)
model=classifier.fit_predict(X)




```

```python
model
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['cluster_group']=model
```

```python
supervised
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(
    data=supervised,
    x=supervised.columns[3],
    y=supervised.columns[4],
    hue=supervised.columns[5]
)    
    
```

```python

n_clusters_ = len(set(model)) - (1 if -1 in model else 0)
n_noise_ = list(model).count(-1)

print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
```
