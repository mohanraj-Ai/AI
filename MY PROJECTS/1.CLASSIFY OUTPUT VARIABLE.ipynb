```python
df=pd.read_csv("feature_selected_data.csv")
```

```python
df
```

```python
def efficiency_class(x):
    if x > 18:
        return "High"
    elif x >= 15:
        return "Medium"
    else:
        return "Low"

df["fuel_efficiency_class"] = df["fuel_efficiency_km_per_l"].apply(efficiency_class)
```

```python
df
```

```python
df=df.drop(columns=['fuel_efficiency_km_per_l'])
```

```python
df
```

```python
df['fuel_efficiency_class'].value_counts()
```

```python
import pandas as pd
```

```python
df_high = df[df['fuel_efficiency_class'] == 'High']
df_medium = df[df['fuel_efficiency_class'] == 'Medium']
df_low = df[df['fuel_efficiency_class'] == 'Low']
```

```python
n_samples = len(df_low)  # 986,429
```

```python
df_high_down = df_high.sample(n=n_samples, random_state=0)
df_medium_down = df_medium.sample(n=n_samples, random_state=0)
```

```python
df_balanced = pd.concat([df_high_down, df_medium_down, df_low], axis=0)
```

```python
df_balanced = df_balanced.sample(frac=1, random_state=0).reset_index(drop=True)

print(df_balanced['fuel_efficiency_class'].value_counts())
```

```python
df_balanced.to_csv("clean_data.csv", index=False)
```
