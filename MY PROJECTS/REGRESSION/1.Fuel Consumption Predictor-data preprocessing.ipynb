```python
import pandas as pd
```

```python
df=pd.read_csv("Reduced-Data.csv")
```

```python
df 
```

```python
df['Ft'] = df['Ft'].str.lower().str.strip()
```

```python
df_pd = df[df['Ft'].isin(['petrol', 'diesel'])]
```

```python
df_pd
```

```python
df_pd = df_pd.drop(columns=['z (Wh/km)', 'Electric range (km)'], errors='ignore')

```

```python
df_pd
```

```python
rename_dict = {
    'm (kg)': 'vehicle_mass_kg',
    'Mt': 'test_mass_kg',
    'ec (cm3)': 'engine_capacity_cc',
    'ep (KW)': 'engine_power_kw',
    'Ewltp (g/km)': 'co2_wltp_g_km',
    'Erwltp (g/km)': 'co2_real_world_g_km',
    'Ft': 'fuel_type',
    'Fm': 'fuel_mode',
    'Fuel consumption': 'fuel_consumption'
}
```

```python
df_pd = df_pd.rename(columns=rename_dict)
```

```python
df_pd
```

```python
# Drop the column 'r'
df_pd = df_pd.drop(['r'], axis=1)
```

```python
df_pd
```

```python
# Check missing values for each column
print(df_pd.isnull().sum())
```

```python
# Remove leading/trailing spaces from all column names
df_pd.columns = df_pd.columns.str.strip()

# Now drop rows where the target/output is missing
df_pd = df_pd.dropna(subset=['Fuel consumption'])

# Check remaining missing values
print(df_pd.isnull().sum())
```

```python
# Rename the column
df_pd.rename(columns={'Fuel consumption': 'Fuel_consumption'}, inplace=True)

# Verify the change
print(df_pd.columns)
```

```python
import pandas as pd

# 1. Strip column names to remove any spaces
df_pd.columns = df_pd.columns.str.strip()

#

# 3. Impute missing numerical features with median
numerical_features_to_impute = ['vehicle_mass_kg', 'test_mass_kg', 'co2_wltp_g_km',
                                'engine_capacity_cc', 'engine_power_kw', 'co2_real_world_g_km']

for col in numerical_features_to_impute:
    if col in df_pd.columns:
        df_pd[col].fillna(df_pd[col].median(), inplace=True)

# 4. Check that there are no missing values in target
print(f"Missing values in target: {df_pd['Fuel_consumption'].isnull().sum()}")

# 5. Verify remaining missing values
print(df_pd.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_features = ['vehicle_mass_kg', 'test_mass_kg', 'co2_wltp_g_km',
                      'engine_capacity_cc', 'engine_power_kw', 'co2_real_world_g_km', 'Fuel_consumption']

plt.figure(figsize=(15,10))
for i, col in enumerate(numerical_features, 1):
    plt.subplot(3,3,i)
    sns.boxplot(y=df_pd[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
numerical_features = ['vehicle_mass_kg', 'test_mass_kg', 'co2_wltp_g_km',
                      'engine_capacity_cc', 'engine_power_kw', 'co2_real_world_g_km', 'Fuel_consumption']

outlier_counts = {}

for col in numerical_features:
    Q1 = df_pd[col].quantile(0.25)
    Q3 = df_pd[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count outliers
    count = ((df_pd[col] < lower_bound) | (df_pd[col] > upper_bound)).sum()
    outlier_counts[col] = count

# Print number of outliers per column
for col, count in outlier_counts.items():
    print(f"{col}: {count} outliers")
```

```python
# List of all numerical columns including target
numerical_cols = ['vehicle_mass_kg', 'test_mass_kg', 'co2_wltp_g_km',
                  'engine_capacity_cc', 'engine_power_kw', 'co2_real_world_g_km',
                  'Fuel_consumption']

# Cap outliers using IQR method
for col in numerical_cols:
    Q1 = df_pd[col].quantile(0.25)
    Q3 = df_pd[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_pd[col] = df_pd[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Save preprocessed dataset to CSV
df_pd.to_csv('preprocessed_vehicle_data.csv', index=False)

print("Dataset saved as 'preprocessed_vehicle_data.csv'")
```
