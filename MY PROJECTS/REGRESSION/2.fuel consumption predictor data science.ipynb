```python
import pandas as pd
```

```python
df_pd=pd.read_csv("preprocessed_vehicle_data.csv")
```

```python
df_pd.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df_pd['fuel_efficiency_km_per_l'], bins=50, kde=True)
plt.title('Distribution of Fuel Consumption')
plt.xlabel('Fuel Consumption')
plt.show()
```

```python
# Take a small random sample for visualization
df_sample = df_pd.sample(n=200000, random_state=42)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df_sample['vehicle_mass_kg'],
            df_sample['fuel_efficiency_km_per_l'],
            alpha=0.3)

plt.xlabel('Vehicle Mass (kg)')
plt.ylabel('Fuel Efficiency (km/L)')
plt.title('Vehicle Mass vs Fuel Efficiency')
plt.show()
```

```python
plt.figure()
plt.scatter(df_sample['engine_capacity_cc'],
            df_sample['fuel_efficiency_km_per_l'],
            alpha=0.3)

plt.xlabel('Engine Capacity (cc)')
plt.ylabel('Fuel Efficiency (km/L)')
plt.title('Engine Capacity vs Fuel Efficiency')
plt.show()
```

```python
plt.figure()
plt.scatter(df_sample['engine_power_kw'],
            df_sample['fuel_efficiency_km_per_l'],
            alpha=0.3)

plt.xlabel('Engine Power (kW)')
plt.ylabel('Fuel Efficiency (km/L)')
plt.title('Engine Power vs Fuel Efficiency')
plt.show()
```

```python
plt.figure()
plt.scatter(df_sample['co2_real_world_g_km'],
            df_sample['fuel_efficiency_km_per_l'],
            alpha=0.3)

plt.xlabel('CO₂ Real World (g/km)')
plt.ylabel('Fuel Efficiency (km/L)')
plt.title('CO₂ Emissions vs Fuel Efficiency')
plt.show()
```

```python
plt.figure()
df_sample.boxplot(column='fuel_efficiency_km_per_l',
                  by='fuel_type')

plt.xlabel('Fuel Type')
plt.ylabel('Fuel Efficiency (km/L)')
plt.title('Fuel Type vs Fuel Efficiency')
plt.suptitle('')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.hexbin(
    df_pd['vehicle_mass_kg'],
    df_pd['fuel_efficiency_km_per_l'],
    gridsize=50,
    mincnt=1
)

plt.xlabel('Vehicle Mass (kg)')
plt.ylabel('Fuel Efficiency (km/L)')
plt.title('Density: Vehicle Mass vs Fuel Efficiency')
plt.colorbar(label='Data Density')
plt.show()
```

```python
import pandas as pd

bins = pd.cut(df_pd['engine_capacity_cc'], bins=30)
grouped = df_pd.groupby(bins)['fuel_efficiency_km_per_l'].mean()

plt.figure(figsize=(8,5))
grouped.plot()

plt.xlabel('Engine Capacity (cc)')
plt.ylabel('Avg Fuel Efficiency (km/L)')
plt.title('Engine Capacity vs Average Fuel Efficiency')
plt.grid(True)
plt.show()
```

```python
bins = pd.cut(df_pd['engine_power_kw'], bins=30)
grouped = df_pd.groupby(bins)['fuel_efficiency_km_per_l'].mean()

plt.figure(figsize=(8,5))
grouped.plot()

plt.xlabel('Engine Power (kW)')
plt.ylabel('Avg Fuel Efficiency (km/L)')
plt.title('Engine Power vs Average Fuel Efficiency')
plt.grid(True)
plt.show()
```
