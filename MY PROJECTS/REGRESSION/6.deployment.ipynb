```python
import pickle
import pandas as pd
import numpy as np

# 1. Load the trained model
model = pickle.load(open('final_model_rf.sav', 'rb'))

# 2. Hardcoded Scaling Parameters (Mean and Std Dev from training data)
TRAINING_MEAN = np.array([1508.43, 1637.21, 1618.54, 101.45, 0.76])
TRAINING_STD  = np.array([309.82, 332.14, 502.81, 39.47, 0.42])

# 3. Define the specifications for 5 different cars
cars_data = {
    "Swift": [900, 1050, 1197, 61, 1],
    "City": [1150, 1300, 1498, 89, 1],
    "Civic": [1300, 1450, 1597, 88, 1],
    "Corolla": [1350, 1500, 1798, 103, 1],
    "Polo": [1100, 1250, 999, 81, 1]
}

results = []

# 4. Process each car and get predictions
for car_name, specs in cars_data.items():
    # Convert list to numpy array
    raw_features = np.array(specs)
    
    # Standardise: (x - mean) / std
    standardised_features = (raw_features - TRAINING_MEAN) / TRAINING_STD
    
    # Prepare input for model (matching feature names from training)
    feature_names = ['vehicle_mass_kg', 'test_mass_kg', 'engine_capacity_cc', 'engine_power_kw', 'fuel_type_petrol']
    input_df = pd.DataFrame([standardised_features], columns=feature_names)
    
    # Predict
    prediction = model.predict(input_df)[0]
    
    # Append only the Name and the Result
    results.append({
        "Car Name": car_name,
        "Predicted Fuel Efficiency (km/l)": round(prediction, 2)
    })

# 5. Create and display the final table
output_table = pd.DataFrame(results)
print(output_table.to_string(index=False))
```
