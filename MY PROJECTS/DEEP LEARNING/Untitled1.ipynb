```python
import numpy as np
from tensorflow.keras.models import load_model

from tensorflow.keras.preprocessing import image
```

```python
model = load_model("tire_defect_model.h5")
print("Model loaded successfully!")
```

```python
img_path = "test.jpg"

img = image.load_img(img_path, target_size=(224, 224))  # Use same size used during training
img_array = image.img_to_array(img)
img_array = img_array / 255.0
img_array = np.expand_dims(img_array, axis=0)
```

```python
prediction = model.predict(img_array)
print("Raw prediction:", prediction)
```

```python
prediction = model.predict(img_array)
prob = prediction[0][0]

if prob > 0.5:
    print(f"рџџў Good Tyre ({prob*100:.2f}% confidence)")
else:
    print(f"рџ”ґ Defective Tyre ({(1-prob)*100:.2f}% confidence)")
```
