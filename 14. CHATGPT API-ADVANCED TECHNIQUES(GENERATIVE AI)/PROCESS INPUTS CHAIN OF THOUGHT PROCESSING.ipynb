```python
import os
from google import genai
from google.genai import types

def get_gemini_completion(user_query):
    # The client will now find the key automatically from the environment
    client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
    
    delimiter = "####"
    
    system_instruction = f"""
    Follow these steps to answer the customer queries.
    The customer query will be delimited with four hashtags, i.e. {delimiter}. 

    Step 1:{delimiter} Decide if the user is asking about a specific product.
    Step 2:{delimiter} Check if the product is in this list:
    1. TechPro Ultrabook (TP-UB100) - 13.3", 8GB RAM, 256GB SSD
    2. BlueWave Gaming Laptop (BW-GL200) - 15.6", 16GB RAM, 512GB SSD, RTX 3060
    3. PowerLite Convertible (PL-CV300) - 14" Touch, 8GB RAM
    4. TechPro Desktop (TP-DT500) - i7, 16GB RAM, 1TB HDD
    5. BlueWave Chromebook (BW-CB100) - 11.6", 4GB RAM

    Step 3:{delimiter} List user assumptions.
    Step 4:{delimiter} Verify assumptions against the data above.
    Step 5:{delimiter} Respond to the user politely, correcting errors if needed.

    Format:
    Step 1:{delimiter} <reasoning>
    Step 2:{delimiter} <reasoning>
    Step 3:{delimiter} <reasoning>
    Step 4:{delimiter} <reasoning>
    Response to user:{delimiter} <final answer>
    """

    try:
        response = client.models.generate_content(
            model="gemini-3.1-flash-lite-preview",
            contents=f"{delimiter}{user_query}{delimiter}",
            config=types.GenerateContentConfig(
                system_instruction=system_instruction,
                temperature=0,
            ),
        )
        return response.text
    except Exception as e:
        return f"Error during generation: {e}"

# Test the setup
query = "Does the BlueWave Chromebook have more RAM than the TechPro Ultrabook?"
print(get_gemini_completion(query))
```
