```python
import google.generativeai as genai
import gradio as gr
import os
from dotenv import load_dotenv

# 1. Load the security credentials from the .env file
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")

if not api_key:
    raise ValueError("System Error: GOOGLE_API_KEY not found in .env file.")

# 2. Configure the API
genai.configure(api_key=api_key)

# 3. Initialize the Model (2026 Standard)
# 3. Initialize the Model (2026 Standard)
model = genai.GenerativeModel(
    model_name="gemini-2.5-flash", 
    system_instruction="You are a Senior Automotive Service Advisor and Technical Expert. Provide precise, safety-oriented diagnostic advice. Use professional engineering terminology but remain accessible. Lead with safety warnings for critical systems like brakes or fuel"
)

# 4. Start a chat session for conversation memory
chat_session = model.start_chat(history=[])

def CustomGemini(user_input):
    try:
        # Using the session ensures the bot remembers previous messages
        response = chat_session.send_message(user_input)
        return response.text
    except Exception as e:
        return f"Service Error: {str(e)}"

# 5. Launch the Gradio Interface
# Updated Gradio Interface for Automotive Context
demo = gr.Interface(
    fn=CustomGemini, 
    inputs=gr.Textbox(
        label="Diagnostic Input (e.g., 'Engine knocking' or 'ABS light on')", 
        placeholder="Describe the vehicle symptoms, noises, or warning lights you are seeing..."
    ), 
    outputs="markdown",
    title="🔧 AI Car Service Advisor",
    description="Professional automotive diagnostics and maintenance guidance powered by Gemini 2.5."
)

if __name__ == "__main__":
    demo.launch(share=True)
```
