```python
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_core.tools import tool
from langgraph.prebuilt import create_react_agent
from langgraph.checkpoint.memory import MemorySaver # Replacement for Memory
from langchain_core.messages import HumanMessage

# 🔐 Load API keys
load_dotenv(".env")

# 🔸 Initialize Gemini LLM
# This model is optimized for the tool-calling logic of ReAct agents
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite", 
    temperature=0
)

# ✅ Tool 1: Simple QA (Functional Replacement for LLMChain)
@tool
def simple_qa(question: str) -> str:
    """Answers factual questions clearly and without hallucination."""
    response = llm.invoke(f"Answer clearly: {question}")
    return response.content

# ✅ Tool 2: Web Search (Tavily)
search_tool = TavilySearchResults(max_results=3)

# 🔧 Agent Configuration
tools = [simple_qa, search_tool]

# 🧠 Memory Setup (Modern Checkpointing)
# Instead of a variable, memory is now a 'checkpointer' that saves state
memory = MemorySaver()

# 🚀 Initialize the Agent (Modern version of initialize_agent)
# Type: ZERO_SHOT_REACT equivalent
agent_executor = create_react_agent(llm, tools, checkpointer=memory)

# 🏁 Running the agent
# To use memory, we define a thread_id (like a session ID)
config = {"configurable": {"thread_id": "session_123"}}

query = "Summarize LangChain in 2 lines, then tell me who created it."
result = agent_executor.invoke(
    {"messages": [HumanMessage(content=query)]}, 
    config=config
)

print("\n🤖 Agent Response:", result["messages"][-1].content)
```
