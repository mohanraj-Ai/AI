```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.prebuilt import create_react_agent
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage, AIMessage

load_dotenv(".env")

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0
)

@tool
def simple_qa(question: str) -> str:
    """Answer factual questions clearly."""
    return llm.invoke(f"Answer clearly: {question}").content

agent = create_react_agent(llm, tools=[simple_qa])

chat_history = []

def run_agent_with_buffer(user_input):
    chat_history.append(HumanMessage(content=user_input))
    result = agent.invoke({"messages": chat_history})
    ai_msg = result["messages"][-1]
    chat_history.append(AIMessage(content=ai_msg.content))
    return ai_msg.content
```

```python
print("Q1:")
print(run_agent_with_buffer("What is LangChain?"))

print("\nQ2 (follow-up):")
print(run_agent_with_buffer("Who created it?"))

print("\nQ3 (context test):")
print(run_agent_with_buffer("Explain it simply again."))
```

```python
def run_agent_no_memory(user_input):
    result = agent.invoke({
        "messages": [HumanMessage(content=user_input)]
    })
    return result["messages"][-1].content
```

```python
print(run_agent_no_memory("What is LangChain?"))
print(run_agent_no_memory("Who created it?"))
```
