```python
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
# ✅ FIX: Import Tool from langchain_core instead of langchain.agents
from langchain_core.tools import Tool 

# 🔐 Load API key
load_dotenv(dotenv_path=".env")

# 🔸 Step 1: Initialize Gemini 2.5 Flash
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash", 
    temperature=0
)

# 🔸 Step 2: Create the Chain
prompt = ChatPromptTemplate.from_template("Answer clearly: {question}")
chain = prompt | llm | StrOutputParser()

# 🔸 Step 3: Wrap the Chain as a Tool
# We use chain.invoke because it's an LCEL runnable
qa_tool = Tool(
    name="Gemini_QA_Tool",
    func=chain.invoke,
    description="Useful for answering general questions using Gemini 2.5 Flash"
)

# 🚀 Step 4: Use the Tool
query = "What is LangGraph in LangChain?"
# Because our chain expects a dictionary, we pass the dict to tool.run
answer = qa_tool.run({"question": query})

# 🖨️ Output
print("\nUser Question:", query)
print("-" * 30)
print("Tool Answer:", answer)
```
