```python
age= int(input("Enter the Age:"))
```

```python
def AgeCategory():
        if(age<18):
            print("Children")
            cate="Children"
        elif(age<35):
            print("Adult")
            cate="Adult"
        elif(age<59):
            print("citizen")
            cate="citizen"
        else:
            print("Senior Citizen")
            cate="Senior Citizen"
        return cate
```
