#2.	Scenario: A digital locker requires users to enter a numerical passcode. As part of a security feature, 
#the system checks the sum of the digits of the passcode.
#Write logic to compute the sum of the digits of a given number


```python
number = int(input("Enter the passcode: "))
digit_sum = 0

while number > 0:
    digit = number % 10
    digit_sum += digit
    number //= 10

print("Sum of digits:", digit_sum)
```

```python
#3.	Scenario: In a secure login system, certain features are enabled only for users with prime-numbered user IDs.
#Write logic to check if a given number is prime
```

```python
num = int(input("Enter the number: "))
if num > 1:
    # Check for factors
    for i in range(2, num):
        if (num % i) == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        # This only runs if the loop finished without finding a divisor
        print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
4.#A scientist is working on permutations and needs to calculate the factorial of numbers frequently.
 #Write logic to find the factorial of a given number using recursion
```

```python
def get_factorial_while(n):
    result = 1
    while n > 1:
        result *= n  # Multiply current result by n
        n -= 1       # Decrease n by 1 for the next loop
    return result

print(f"The factorial of 5 is: {get_factorial_while(5)}")
```

#5.Scenario: A unique lottery system assigns ticket numbers where only Armstrong numbers win the jackpot.
#Write logic to check whether a given number is an Armstrong number


```python
number = int(input("Enter the lottery ticket number: "))
original = number
digits = len(str(number))
armstrong_sum = 0

while number > 0:
    digit = number % 10
    armstrong_sum += digit ** digits
    number //= 10

if armstrong_sum == original:
    print(original, "is an Armstrong number")
else:
    print(original, "is NOT an Armstrong number")
```

```python
#6.A password manager needs to strengthen weak passwords by swapping the first and last characters of user-generated passwords.
#Write logic to perform this operation on a given string.
```

```python
password = input("Enter password: ")

if len(password) < 2:
    print("Password after swap:", password)
else:
    swapped = password[-1] + password[1:-1] + password[0]
    print("Password after swap:", swapped)
```

#7.	Scenario: A mobile payment app uses a simple checksum validation where reversing a transaction ID helps detect fraud.
 Write logic to take a number and return its reverse.



```python
number = int(input("Enter transaction ID: "))
reverse_num = 0

while number > 0:
    digit = number % 10
    reverse_num = reverse_num * 10 + digit
    number //= 10

print("Reversed transaction ID:", reverse_num)
```

```python
#8.Scenario: A low-level networking application requires decimal numbers to be converted into binary format before transmission.
#Write logic to convert a given decimal number into its binary equivalent.

```

```python
decimal = int(input("Enter decimal number: "))
binary = ""

if decimal == 0:
    binary = "0"
else:
    while decimal > 0:
        remainder = decimal % 2
        binary = str(remainder) + binary
        decimal //= 2

print("Binary equivalent:", binary)
```

```python
#9.	Scenario: A text-processing tool helps summarize articles by identifying the most significant words.
#Write logic to find the longest word in a sentence.

```

```python
sentence = input("Enter a sentence: ")
words = sentence.split()

# The 'key=len' tells Python to compare the items by their length 
# rather than their alphabetical order.
longest_word = max(words, key=len)

print("Longest word:", longest_word)
```

```python
#:10.A plagiarism detection tool compares words from different documents and checks if they are anagrams (same characters but different order).
#Write logic to check whether two given strings are anagrams.
```

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

# Convert to lowercase and remove spaces
str1 = str1.replace(" ", "").lower()
str2 = str2.replace(" ", "").lower()

# Check lengths
if len(str1) != len(str2):
    print("Not anagrams")
else:
    # Sort and compare
    if sorted(str1) == sorted(str2):
        print("Anagrams")
    else:
        print("Not anagrams")
```
