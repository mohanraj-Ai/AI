```python
import pandas as pd
```

```python
df = pd.read_csv("mabel.txt", header=None, on_bad_lines='skip', encoding='utf-8')
```

```python
df
```

```python
df=df.drop(0)
df.columns=['Date','Chat']
Message=df["Chat"].str.split("-",n=1,expand=True)
df["Time"]=Message[0]
Message1=Message[1].str.split(":",n=1,expand=True)
df["Name"]=Message1[0]
df["Chat"]=Message1[1]
df=df[["Date","Time","Name","Chat"]]
df

```

```python
import pandas as pd

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%y')

```

```python
df
```

```python
datewise_count = df.groupby(['Date', 'Name']).size().reset_index(name='Message_Count')
pivot_df = datewise_count.pivot(index='Date', columns='Name', values='Message_Count').fillna(0)
```

```python
pivot_df
```

```python
datewise_count
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(15,7))

# Plot bar chart
pivot_df.plot(kind='bar', stacked=True, width=0.8, figsize=(15,7))

plt.title("Datewise Message Count per Person")
plt.xlabel("Date")
plt.ylabel("Number of Messages")
plt.xticks(rotation=45)  # Rotate date labels
plt.legend(title="Sender")
plt.tight_layout()
plt.show()
```
